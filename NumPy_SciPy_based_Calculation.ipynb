```python
import numpy as np
import scipy as sp
```

```python
A = np.array([
    [2,  1, -1,  3],
    [4, -2,  5,  1],
    [-1, 3,  2, -4],
    [3, -1,  4,  2]
])

print(A)
```

```python
b = np.array([[5, 16, -2, 10]])
b = b.T

print(b)
```

```python
Xnp = np.linalg.solve(A, b)
Xsp = sp.linalg.solve(A, b)

Xnp = Xnp.T
Xsp = Xsp.T

print("Xnp = ", Xnp)
print("Xsp = ", Xsp)
```

```python
P, L, U = sp.linalg.lu(A)

#for debugging
print("L = ")
print(L)
print("U = ")
print(U)
print("P = ")
print(P)
```

```python
LU, P = sp.linalg.lu_factor(A)
XscpLU = sp.linalg.lu_solve((LU, P), b)

#for debugging
print("LU = ")
print(LU)
print("P = ")
print(P)
print("XscpLU = ")
print(XscpLU)
```
